```r
pp<-read.csv('PPDAY1.csv',header=TRUE,row.names = 1)
```

```r
library(factoextra)
library(ggplot2)
set.seed(1234)
```

```r
fviz_nbclust(pp, kmeans, method = "wss")
```

```r
km.res <- kmeans(pp,8)
km.res$cluster
```

```r
c<-km.res$cluster
```

```r
library(Rtsne)
```

```r
tsne_out <- Rtsne(
  pp,
  dims = 2,
  pca = TRUE,
  perplexity = 10,
  theta = 0.0,
  max_iter = 100
)
```

```r
x<-tsne_out$Y[,1]
```

```r
y<-tsne_out$Y[,2]
```

```r
to<-data.frame(x,y,c)
```

```r
to$index=rownames(to)
```

```r
write.csv(to,'T-snee-Day1.csv')
```
